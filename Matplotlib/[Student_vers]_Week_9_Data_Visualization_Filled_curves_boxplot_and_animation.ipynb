First: Plot $\sin(2 \pi x)$ where $x\in [0,2]$

```python
import matplotlib.pyplot as plt
import numpy as np

def fungsi2(x):
    return np.sin(2*np.pi*x)

x = np.linspace(0, 2, 100)

plt.xlabel('x')
plt.ylabel('y')

plt.grid()

plt.plot(x, fungsi2(x), color='red', linewidth=3)

plt.legend(('sin(2pix)',), loc='upper right')
plt.title('Plot sin(2pix)', fontsize=20)
plt.show()
```

To filled the curve with color from 0 to line of $\sin(2 \pi x)$, the use plt.fill_between(x, 0, y)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2, 0.01)
y = np.sin(2*np.pi*x)

plt.xlabel('x')
plt.ylabel('y')

plt.xlim(min(x)-0.1, max(x)+0.1)
plt.ylim(min(y)-0.1, max(y)+0.1)

plt.grid()

plt.plot(x, y, color='red', linewidth=3)
plt.fill_between(x, 0, y)

plt.show()
```

To filled the curve with color between two curves, use the following syntax



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)

plt.xlabel('x')
plt.ylabel('y')

plt.xlim(min(x)-0.1, max(x)+0.1)
plt.ylim(min(y1)-0.1, max(y1)+0.1)

plt.grid()

plt.plot(x, y1, color='red', linewidth=3)
plt.plot(x, y2, color='green', linewidth=3)
plt.fill_between(x, y2, y1)

plt.show()
```

To filled the curve with color between two curves, where sin > cos, use the following syntax




```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)

plt.xlabel('x')
plt.ylabel('y')

plt.xlim(min(x)-0.1, max(x)+0.1)
plt.ylim(min(y1)-0.1, max(y1)+0.1)

plt.grid()

plt.plot(x, y1, color='red', linewidth=3)
plt.plot(x, y2, color='green', linewidth=3)
plt.fill_between(x, y2, y1, where=y1 >= y2)
 
plt.show()
```

To filled the curve with color between two curves, where sin > cos and masked if cos > 0, use the following syntax

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)

plt.xlabel('x')
plt.ylabel('y')

plt.xlim(min(x)-0.1, max(x)+0.1)
plt.ylim(min(y1)-0.1, max(y1)+0.1)

plt.grid()

plt.plot(x, y1, color='red', linewidth=3)
plt.plot(x, y2, color='green', linewidth=3)
plt.fill_between(x, y2, y1, where=y1 >= y2)
 
plt.show()
```

# Boxplot for statistics

Boxplots are a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”).


*   **median (Q2/50th Percentile)**: the middle value of the dataset.
*   **first quartile (Q1/25th Percentile)**: the middle number between the smallest number (not the “minimum”) and the median of the dataset.
*   **third quartile (Q3/75th Percentile)**: the middle value between the median and the highest value (not the “maximum”) of the dataset.
*   **interquartile range (IQR)**: 25th to the 75th percentile.
*   **whiskers** (shown in blue)
*   **outliers** (shown as green circles)
*   **“maximum”**: Q3 + 1.5*IQR
*   **“minimum”**: Q1 -1.5*IQR
