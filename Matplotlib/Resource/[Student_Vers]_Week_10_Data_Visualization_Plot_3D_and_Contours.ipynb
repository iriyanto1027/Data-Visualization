One of useful toolkit to visualize dataset in 3D provided by Matplotlib is **mplot3D**. 
The mplot3d toolkit adds simple 3D plotting capabilities to matplotlib by supplying an axes object that can create a 2D projection of a 3D scene. The resulting graph will have the same look and feel as regular 2D plots.
The interactive backends also provide the ability to rotate and zoom the 3D scene. One can rotate the 3D scene by simply clicking-and-dragging the scene. Zooming is done by right-clicking the scene and dragging the mouse up and down. Note that one does not use the zoom button like one would use for regular 2D plots.

**Another Example**

We want to visualize the contour / levelset of the function. 

$f(x,y) = \sin(x)^2 + \cos(10 + y) \cos(x)$  for $(x,y) \in [-3,3] \times [-3,3]$

The plot in the right side is a nice contour plot using funcion ***contour***, but the spaces between the lines may be a bit distracting. 
Also, it is possible to plot filled contour plot using ***contourf()*** and add the colorbar that is automatically generated by function ***colorbar()***.

```
# contour PLot library
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

# preparing 3D data points using 'meshgrid'. The smoothness depends on delta. 
delta = 0.0025
x = np.arange( , delta)
y = np.arange( , delta)
X, Y = np.meshgrid(x, y)
Z =  

# create the object subplot consist 1x2 plots. 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))

# controling line of levelset for the left plot in the figure
CS = ax1.contour(X, Y, Z, 20,
                 linewidths=np.arange(.5, 4, .5),
                 cmap='RdGy'
                 )
ax1.clabel(CS, fontsize=9, inline=1)
ax1.set_title('Contour plot with contour level lie in the levelset')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

#create a filled contour plot using contourf equipped with the colorbar.
cs = ax2.contourf(X, Y, Z, 20, cmap='RdGy')
fig.colorbar(cs)
ax2.set_title('Filled contour plot with the colorbar')
```
